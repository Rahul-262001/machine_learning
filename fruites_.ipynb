```python
import pandas as pd
df=pd.read_csv("fruits.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
le = LabelEncoder()
mm = MinMaxScaler()
df["Class"]=le.fit_transform(df["Class"])
for i in ["Weight","Size"]:
    df[i] = mm.fit_transform(df[[i]])
```

```python
df.head()
```

```python
y = df["Class"]
x = df[["Weight","Size"]]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,shuffle=True,random_state=1,train_size=0.8)
```

```python
from sklearn.svm import LinearSVC as clf
clf = clf(C=10)
clf.fit(xtrain,ytrain)
```

```python
ypred = clf.predict(xtest)
from sklearn.metrics import accuracy_score as a_s
print(a_s(ytest,ypred))
```

# Train and Test ACCURACY

```python
print(clf.score(xtrain,ytrain))
print(clf.score(xtest,ytest))
```

```python
from mlxtend.plotting import plot_decision_regions as pdr
import numpy as np
pdr(np.array(xtrain),np.array(ytrain),clf)
```

```python
import pandas as pd
df=pd.read_csv("BMI.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
le = LabelEncoder()
mm = MinMaxScaler()
df["Gender"]=le.fit_transform(df["Gender"])
df[['Height','Weight']]=mm.fit_transform(df[['Height','Weight']])
```

```python
y = df["Index"]
x = df[['Height','Weight','Gender']]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,shuffle=True,random_state=1,train_size=0.8)
```

```python
from sklearn.svm import LinearSVC as clf
clf = clf()
clf.fit(xtrain,ytrain)
ypred = clf.predict(xtest)
from sklearn.metrics import accuracy_score as a_s
print(a_s(ytest,ypred))
```

```python
print(clf.score(xtrain,ytrain))
print(clf.score(xtest,ytest))
```

```python
from mlxtend.plotting import plot_decision_regions as pdr
import numpy as np
pdr(np.array(xtrain),np.array(ytrain),clf)
```
